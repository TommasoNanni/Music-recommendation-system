```python
import pandas as pd
import numpy as np

data_genres = pd.read_csv('../data/genres.csv')
data_tracks = pd.read_csv('../data/tracks.csv')
data_features = pd.read_csv('../data/echonest.csv')
```

```python
# drop redundant cols
data_clean = data_tracks.drop(columns=["album", "album.1", "album.3", "album.4", "album.5", "album.6", "album.8", "album.9", "album.12","artist.1", "artist.2", "artist.4", "artist.5", "artist.6", "artist.8", "artist.10", "artist.11", "artist.13", "artist.14", "artist.15", "artist.16", "set", "set.1", "track", "track.1","track.2", "track.4", "track.6", "track.7", "track.8", "track.10", "track.11", "track.13", "track.15", "track.16", "track.17", "track.18"])
```

```python
# rename columns
data_clean.columns = [
    "track_id", "album_date_release", "album_listens", "album_title", "tracks_in_album",
    "artist_year", "artist_bio", "artist_id", "artist_location", "artist_name",
    "track_date_release", "duration", "genres_ids", "language", "listens", "track_title",
]
```

```python
# drop first 2 rows containing headers
data_clean = data_clean.iloc[2:].reset_index(drop=True)
```

```python
# separate artist and tracks data
df_artist = data_clean[["artist_id", "artist_name", "artist_year", "artist_location", "artist_bio"]].drop_duplicates().reset_index(drop=True)
df_tracks = data_clean.drop(columns=["artist_name", "artist_year", "artist_location", "artist_bio"]).reset_index(drop=True)
```

```python
df_artist.to_csv('../data/clean_artist.csv', index=False)
df_tracks.to_csv('../data/clean_tracks.csv', index=False)
data_genres.to_csv('../data/clean_genres.csv', index=False)
```
